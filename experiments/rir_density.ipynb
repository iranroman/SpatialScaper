```python
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append('../data/')
import sofa_utils
```

```python
ELE_RANGE = (-.4 * np.pi, .4 * np.pi)
AZI_RANGE = (-np.pi, np.pi)
```

```python
#TAU_DENSITY
n_ele = 9
n_azi = 360

eles = np.linspace(ELE_RANGE[0], ELE_RANGE[1], n_ele)
azis = np.linspace(AZI_RANGE[0], AZI_RANGE[1], n_azi)


fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')


xs = []
ys = []
zs = []
for ele in eles:
    for azi in azis:
        xs.append(np.cos(azi)*np.cos(ele))
        ys.append(np.sin(azi)*np.cos(ele))
        zs.append(np.sin(ele))

        
ax.scatter(xs,ys,zs, marker='.', alpha=0.2)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```
